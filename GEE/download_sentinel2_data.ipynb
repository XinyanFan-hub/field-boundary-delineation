```python
def maskS2clouds(image): 
    qa = image.select('SCL');
    shadowBitMask = 1 << 3
    cloud1BitMask = 1 << 8
    cloud2BitMask = 1 << 9
    cirrusBitMask = 1 << 10
    mask = qa.bitwiseAnd(shadowBitMask).eq(0) and qa.bitwiseAnd(cloud1BitMask).eq(0) and qa.bitwiseAnd(cloud2BitMask).eq(0) and qa.bitwiseAnd(cirrusBitMask).eq(0)
           
    return image.updateMask(mask)
```
