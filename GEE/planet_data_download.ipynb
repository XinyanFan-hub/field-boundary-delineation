```python
Month2021 =  ee.ImageCollection('projects/planet-nicfi/assets/basemaps/asia')\
              .filterDate('2021-01-01','2021-12-31')\
              .filterMetadata('cadence', 'equals','monthly');
Month2017 =  ee.ImageCollection('projects/planet-nicfi/assets/basemaps/asia')\
              .filterDate('2017-01-01','2017-12-31')\
              .filterMetadata('cadence', 'equals','monthly');
Month2018 =  ee.ImageCollection('projects/planet-nicfi/assets/basemaps/asia')\
              .filterDate('2018-01-01','2018-12-31')\
              .filterMetadata('cadence', 'equals','monthly');
Month2019 =  ee.ImageCollection('projects/planet-nicfi/assets/basemaps/asia')\
              .filterDate('2019-01-01','2019-12-31')\
              .filterMetadata('cadence', 'equals','monthly');
Month2020 =  ee.ImageCollection('projects/planet-nicfi/assets/basemaps/asia')\
              .filterDate('2020-01-01','2020-12-31')\
              .filterMetadata('cadence', 'equals','monthly');
Month2022 =  ee.ImageCollection('projects/planet-nicfi/assets/basemaps/asia')\
              .filterDate('2022-01-01','2022-12-31')\
              .filterMetadata('cadence', 'equals','monthly');
Month2021_planet = Month2021.median().clip(geometry);  
Month2017_planet = Month2017.median().clip(geometry);  
Month2018_planet = Month2018.median().clip(geometry);  
Month2019_planet = Month2019.median().clip(geometry);  
Month2020_planet = Month2020.median().clip(geometry);  
Month2022_planet = Month2022.median().clip(geometry);  
collection = ee.ImageCollection([Month2021_planet, Month2017_planet, Month2018_planet, Month2019_planet, Month2020_planet, Month2022_planet])
       
```
